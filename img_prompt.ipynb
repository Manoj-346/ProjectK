```python
from transformers import pipeline
```

```python
image_captioner = pipeline("image-to-text", model='Salesforce/blip-image-captioning-large',device=0)
```

```python
image = r'/content/thala.jpg'
caption = image_captioner(image)
caption[0].get('generated_text')
```
