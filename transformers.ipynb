```python
qa = pipeline('question-answering')
```

```python
c = '''Water is essential for all known forms of life. It covers about 71% of the Earth's surface.'''
```

```python
q = "How much of the Earth's surface is covered by water?"
```

```python
result = qa({'question': q, 'context': c})
print(result)
```
