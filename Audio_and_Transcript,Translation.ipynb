```python
from gtts import gTTS
from IPython.display import Audio
```

```python
from pydub import AudioSegment #mp3 format to .wav
mp3_file = "/content/harsha.mp3"
wav_file = "/content/harsha.wav"
audio = AudioSegment.from_mp3(mp3_file)
audio.export(wav_file, format="wav")
```

```python
import speech_recognition as sr
reco = sr.Recognizer()
with sr.AudioFile("/content/harsha.wav") as source:
    audio_data = reco.record(source)
    print("Transcript : ",reco.recognize_google(audio_data))
```

```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
ip = "/content/hw1.jpg"
plt.imshow(Image.open(ip))
plt.axis("off")
```

```python
img = Image.open(ip)
text = pytesseract.image_to_string(img)
print("Extracted text:",text)
```

```python
tts = gTTS(text=text,lang="en")
```

```python
audio_filename = "img.png.mp3"
```

```python
tts.save(audio_filename)
```

```python
Audio(audio_filename)
```
